```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
import csv
unfilte_linklist=[]
url_list=[]
```

```python
def get_unclean_link(url):
    soup = BeautifulSoup(requests.get(url).text, 'lxml')
    #soup is bs4.BeautifulSoup
    soup1 = soup.select("div > p.media-heading")
    #soup1 is a bs4 tag list
    for i in soup1:
        if i in unfilte_linklist:    
            break #if already in list pass 
        else:
            unfilte_linklist.append(i)
        #i is bs4.element.Tag
    return 'unclean_linklist create please filte'
```

```python
url = 'https://forum.babyhome.com.tw/list/1047?year=all&page={}'
for num in range(1,10):
     get_unclean_link(url.format(num))
```

```python
filte_list=[]
def filte_article(unfilte_linklist):
    for link in unfilte_linklist:
        if link.find('a', {'title': re.compile('美白|粉刺|痘|暗沉|緊緻|角質|保濕|防曬|斑|法令紋|皺紋')}) != None:
            filte_list.append(link.find('a', {'title': re.compile('美白|粉刺|痘|暗沉|緊緻|角質|保濕|防曬|斑|法令紋|皺紋')}))
        else:
            pass
    return 'filte_list create'
```

```python
def get_url(filte_list):
    for url in filte_list:
        if url in url_list:  
            break
        else:
            url_list.append(url['href'])
    return 'url_list create'
```

```python
import json
def get_soup(url):
    comment_list=[]
    data = {}
    question1_list=[]
    test2=[]
    test3=[]
    soup = BeautifulSoup(requests.get(url).text, 'lxml')
    soup1 = soup.select_one("div > h1")
    #here only on condition match so we use select_one, if there are more tag match we should use select
    data['title']=soup1.text
    question = soup.select_one("div.thread-inner")
    times = soup.select_one("p.floors.pull-right").text
    question_list=question.select('p')
    question_list.pop()
    question_list.pop()
    for i in question_list:
        if i.text == '\xa0':
            pass
        elif i.text == '':
            pass
        else:
            question1_list.append(i.text)
    data['ask'] = question1_list
    data['times'] = times.replace("版主",'').strip()
    comment = soup.select("div.thread-inner")
    del comment[0]
    #get frist page comment
    for i in comment:
        i.select('button.btn')[0].extract()
        #remove button.btn
        i.select('button.btn')[0].extract()        
        try:
            #remove div class="" type
            i.select('div.')[0].extract()
        except IndexError:
            pass
        try:       
            #remove quote.clearfix need to ask why can use quote.clearfix
            i.select('div.quote.clearfix')[0].extract()
        except IndexError:
            pass
        try:   
            #remove div.edited-comment
            i.select('div.edited-comment')[0].extract()
        except IndexError:
            pass
        
        comment_list.append(i.text.strip())
    
    next_page = soup.select_one('ul.pagination.pull-right')# need to ask why can use like that
    next_url = next_page.find_all('a',{'href':re.compile('/topic/.*')})
    #get after frist page comment
    if '1' in next_page.select('a')[1].text:
        del next_url[-1]
        for url in next_url:
            a='https://forum.babyhome.com.tw'+url['href']
            comment_soup = BeautifulSoup(requests.get(a).text, 'lxml') 
            comment1 = comment_soup.select("div.thread-inner > p")
            for j in comment1:
                if j.text == '\xa0':
                    pass
                elif j.text == '':
                    pass
                else:
                    comment_list.append(j.text)      
    else:
        pass
    
    data['url'] = soup.find('meta', {'content': re.compile('https://forum.babyhome.com.tw/topic/.*')})['content']
    data['comment'] = comment_list
    data_list.append(data)
    return data
```

```python
get_soup('https://forum.babyhome.com.tw/topic/4756696')
```

```python
with open("stock.csv","w") as f:
    w = csv.DictWriter(f, data_list[0].keys())
    w.writeheader()
    w.writerow(data_list[0])
```
