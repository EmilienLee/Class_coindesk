```python


# Blah
import logging
import pandas as pd
import numpy as np
from pprint import pprint
import sys
import os
import requests
import datetime 
import json
import time

# # Charting
import matplotlib.pyplot as plt
from PIL import Image
```

```python
url = 'https://api.bitfinex.com/v2/candles/trade:1D:tBTCUSD/hist?limit=1000'
candles_json = json.loads(requests.get(url).text)

while len(candles_json) < 1000:
    time.sleep(2)
    earliest = (candles_json[len(candles_json)-1])[0] - 1
    url = 'https://api.bitfinex.com/v2/candles/trade:1D:tBTCUSD/hist?limit=1000&end=%s'
    query = url % earliest
    result = json.loads(requests.get(query).text)
    candles_json += result
```

```python
print(candles_json)
```

```python
candles = pd.read_json(json.dumps(candles_json))
candles.rename(columns={0:'date', 1:'open', 2:'close', 3:'high', 4:'low', 5:'volume'}, inplace=True)
candles['date'] = pd.to_datetime( candles['date'], unit='ms' )
candles.set_index(candles['date'], inplace=True)
candles.sort_index(inplace=True)
candles.index = candles.index + pd.DateOffset(hours=0)
```

```python
import tensorflow as tf
import numpy as np
```

```python
# create data
x_data =np.random.rand(100).astype(np.float32)
y_data = x_data*0.1 + 0.3
```

```python
#create tensorflow structure
#it might multidimansion so use captial
Weights=tf.Variable(tf.random_uniform([1],-1.0,1.0))
biases = tf.Variable(tf.zeros([1]))

y = Weights * x_data + biases

loss = tf.reduce_mean(tf.square(y-y_data))#loss function also call cost function
optimizer = tf.train.GradientDescentOptimizer(0.5)#learning rate usually smaller than 1
train = optimizer.minimize(loss) #minimize loss

# init = tf.initialize_all_variables() # tf 
init = tf.global_variables_initializer()  #  #initial variable for tensorflow structure
#create tensorflow end

sess=tf.Session()
#session like a pointer ,to print location the location in the tf structure and run the particular area 
sess.run(init)
# after run initial neural structure activate ***important

for step in range(201):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(Weights), sess.run(biases))
```

## session is a pointer to the whole network picture

```python
# create two matrixes

matrix1 = tf.constant([[3,3]])
matrix2 = tf.constant([[2],
                       [2]])
product = tf.matmul(matrix1,matrix2)#matrix multiply same as np.dot(m1,m2)
### method 1
sess=tf.Session()#it's an object
result= sess.run(product)
print (result)
sess.close()
```
